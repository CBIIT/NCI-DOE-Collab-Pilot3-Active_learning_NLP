```python
import pandas as pd
import pickle
import nltk
import matplotlib.pyplot as plt
from nltk import word_tokenize
import config
```

```python
with open(config.path_to_preprocessed_reports, 'rb') as fh:
    reports = pickle.load(fh)
```

```python
cat_reports = ''
for report in reports:
    cat_reports += report

tokens = word_tokenize(cat_reports)
fdist1 = nltk.FreqDist(tokens)
fdist1.plot(40, cumulative=False,
            title='Top 40 Most Common Words in Corpus')
```

```python
length_rep = list()
for report in reports:
    length_rep.append(len(report))
plt.hist(length_rep, bins=60)
plt.ylabel('# of reports')
plt.xlabel('Number of tokens')
plt.title('Histogram of reports\' length')
plt.show()
```

```python
df_fdist = pd.DataFrame.from_dict(fdist1, orient='index')

df_fdist.columns = ['Frequency']
df_fdist.index.name = 'Term'
df_fdist = df_fdist.sort_values(by=['Frequency'], ascending=False)
df_fdist.to_csv(config.path_to_word_frequency_file)
```

```python
print('Total number of tokens: {}'.format(len(tokens)))
print('Total number of unique tokens: {}'.format(len(set(tokens))))
```

```python
textList = nltk.Text(tokens)
textList.concordance('no')  # find all occurrences of the token in the corpus
```

```python
# print the long words in the corpus
# long words sometimes are erroneous words
cont = 0
for t in tokens:
    if len(t) > 20:
        print(t)
        cont += 1
print(cont)
```
