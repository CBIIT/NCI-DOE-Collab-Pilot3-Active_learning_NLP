```python
import os
import pickle
import matplotlib.pyplot as plt
```

```python
experiment = 'experiment_013'
cont = 0
for folder in os.listdir(experiment):
    for file in os.listdir('{}/{}'.format(experiment, folder)):
        if file.endswith('.pkl'):
            dt_size, perf = pickle.load(open('{}/{}/{}'.format(experiment, folder, file), 'rb'))
            plt.figure(cont)
            plt.plot(dt_size[1:], perf['ResNet-Random'][1:], color='b', label='random')
            plt.plot(dt_size[1:], perf['ResNet-Entropy'][1:], color='r', label='entropy')
            plt.plot(dt_size[1:], perf['ResNet-Abstention'][1:], color='g', label='abstention')
            plt.legend()
            plt.show()
            cont += 1
```
